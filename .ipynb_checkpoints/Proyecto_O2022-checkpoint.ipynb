```python
# Importar librerías
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random
%matplotlib inline
```

```python
# Función para descargar precios de cierre ajustados:
def get_adj_closes(tickers, start_date=None, end_date=None, freq='d'):
    closes = web.YahooDailyReader(symbols=tickers, start=start_date, end=end_date, interval=freq).read()['Adj Close']
    closes.sort_index(inplace=True)
    return closes
```

```python
# Descargar precios y graficar
closes = get_adj_closes(
     ['VTI', 'VGT', 'VCR', 'VFH', 'VHT', 
      'AAPL', 'AMZN', 'GOOGL', 'COST', 'MSFT', 
      'TSLA', 'META', 'TWTR', 'DIS', 'NFLX'],
    start_date="2010-01-01",
    end_date="2022-01-01"
)
returns = closes.pct_change().dropna()
```

```python
# Calculemos las betas
stocks=list(returns.keys())
betas= {}
for i in stocks:
    Y = returns[[i]]
    X = returns[['VTI']]
    model = LinearRegression().fit(X,Y)
    beta_coefficient = model.coef_[0][0]
    betas[i] = [beta_coefficient, Y.mean()[0]]
    # Es el rendimiento promedio de cada activo. (Los rendimientos estaban mensuales acuerdate)
```

```python
# Visualizar beta vs rendimiento 
regression_beta = [betas[i][0] for i in betas]
#regression_beta
returns1 = returns.copy()
returns1 = [betas[i][1] for i in betas]
df = pd.DataFrame()
df['stocks'] = [i for i in betas]
df['beta'] = regression_beta
df['returns'] = returns1
df
```

### Histogramas bootstrap de las betas

```python
returns = closes.pct_change().dropna()
```

```python
activos = list(returns.keys())
activos.remove('VTI')
boot={}
betas = []
for index, activo in enumerate(activos):
    for i in range(10000):
        sample_df = returns.sample(len(returns), replace=True)
        beta = sample_df.cov().values[index+1][0] / np.var(sample_df['VTI'])
        betas.append(beta)
    boot[activo] = betas        
    betas = []
```

```python
#fig, ax = plt.subplots(nrows=7, ncols=2, figsize=(20,10))
for activo in activos:
    df_betas= pd.DataFrame()
    df_betas[f'beta {activo}'] = boot[activo]
    df_betas.hist(bins=100)
```

## R2

```python
activos = list(returns.keys())
r2= {}
for i in activos:
    X = returns[['VTI']]
    Y = returns[[i]]
    model = LinearRegression().fit(X,Y)
    df_copy = returns.copy()
    df_copy['prediction'] = model.predict(X)
    df_copy['MEAN'] = df_copy[i].mean()
    
    var_mean = np.var(df_copy[i] - df_copy.MEAN)
    var_model = np.var(df_copy[i] - df_copy.prediction)
    r2[i] = (var_mean - var_model) / var_mean
```

```python
r2
```

## 2. Selección de portafolios

Una vez elegidos los activos y/o ETFs:

#### 1. Usando los precios históricos de dichos activos hasta el 2015-12-31 (la fecha inicial es un parámetro que ustedes deberán elegir), diseñarán un portafolio en el que (hipotéticamente) habrían invertido durante todo el 2016. Para esto, deberán suponer un coeficiente de aversión al riesgo.


```python
# Descargar precios y graficar
closes = get_adj_closes(
     ['VTI', 'VGT', 'VCR', 'VFH', 'VHT', 
      'AAPL', 'AMZN', 'GOOGL', 'COST', 'MSFT', 
      'TSLA', 'META', 'TWTR', 'DIS', 'NFLX'],
    start_date="2010-01-01",
    end_date="2015-12-31"
)
rt = closes.pct_change().dropna()

# Tasa de Bonos del Tesoro de 2016 a un año
# https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value=2017
rf = 0.0061
```

```python
# Resumen de volatilidad y rendimiento en base anual
df = pd.DataFrame()
df['vol'] = rt.std() * 252 ** 0.5
df['return'] = rt.mean() * 252 
df
```

```python
# Matriz de covarianza
cov = rt.cov()
# Matriz de correlación
corr = rt.corr()
```

```python
# Eind: rendimientos esperados activos individuales
eind = df['return'].values 
# Número de activos
n = len(eind)
eind
```

```python
## Encontrar portafolio de mínima varianza
## Construcción de parámetros
# Sigma: matriz de varianza-covarianza Sigma = S.dot(corr).dot(S)
S = np.diag(df['vol']).astype(float) #.diag crea matriz diagonal
sigma = S.dot(corr).dot(S)
#print(sigma)
```

```python
# Función objetivo
def var(w, sigma):
    return w.T.dot(sigma).dot(w)
```

```python
# Número de activos
n = len(eind)
# Dato inicial
w0 = np.ones(n) / n
# Cotas de las variables
bnds = ((0, 1),) * n
# Restricciones
cons = {'type': 'eq', 'fun': lambda w: w.sum() - 1}
```

```python
from scipy.optimize import minimize
# Portafolio de mínima varianza
minvar = minimize(
    fun=var,
    x0=w0,
    args=(sigma,),
    bounds=bnds,
    constraints=cons
)
```

```python
wminvar = minvar.x
wminvar
```

```python
# Pesos, rendimiento, riesgo y razón de Sharpe del portafolio de mínima varianza
e_mv = wminvar.dot(eind)
s_mv = np.sqrt(wminvar.T.dot(sigma).dot(wminvar))
sharpe_mv= (e_mv-rf)/ s_mv
e_mv, s_mv, sharpe_mv
```

```python
## Encontrar portafolio EMV
# Función objetivo
def menos_rs(w, rf, eind, sigma):
    sp = (w.T.dot(sigma).dot(w))**0.5
    ep = w.dot(eind)
    rs = (ep - rf) / sp
    return -rs
```

```python
# Número de activos
n = len(eind)
# Dato inicial
w0 = np.ones(n) / n
# Cotas de las variables
bnds = ((0, 1),) * n
# Restricciones
cons = {'type': 'eq', 'fun': lambda w: w.sum() - 1}
```

```python
# Portafolio EMV
emv = minimize(
    fun=menos_rs,
    x0=w0,
    args=(rf, eind, sigma),
    bounds=bnds,
    constraints=cons
)
```

```python
wemv_1 = emv.x
wemv_1
```

```python
# Pesos, rendimiento, riesgo y razón de Sharpe del portafolio EMV
e_emv_1 = wemv_1.dot(eind)
s_emv_1 = np.sqrt(wemv_1.dot(sigma).dot(wemv_1))
sharpe_emv_1 = (e_emv_1 - rf) / s_emv_1
e_emv_1, s_emv_1, sharpe_emv_1
```

```python
# Comparación
print('Asi se con optimizado el sharpe')
print(e_emv_1, s_emv_1, sharpe_emv_1)

print('Asi se va con minima varianza')
print(e_mv, s_mv, sharpe_mv)
```

```python
##  Combinación óptima de acuerdo a preferencias
# Coeficientes de aversión al riesgo
# Para gamma
g = np.linspace(1, 15, 15)
```

```python

df = pd.DataFrame({
    'g':g, #aversión al riesgo
    'w_opt': (e_emv_1 - rf)/(g*s_emv_1**2),
    '1-w_opt': 1 - ((e_emv_1 - rf)/(g*s_emv_1**2))
})
# Todas las posibles combinaciones
df # mientras g mas grande mas adverso al riesgo
```

```python
g = np.linspace(5,5,1)
df = pd.DataFrame({
    'g':g, #aversión al riesgo
    'w_opt': (e_emv_1 - rf)/(g*s_emv_1**2),
    '1-w_opt': 1 - ((e_emv_1 - rf)/(g*s_emv_1**2))
})
df
```

Con una aversión al riesgo de 5, el inversionista se apalancaría para de esta manera invertir 190% en acciones.

```python
# Número de activos
n = len(eind)
# Dato inicial
w0 = np.ones(n) / n
# Cotas de las variables
bnds = ((0, 1),) * n
# Restricciones
cons = {'type': 'eq', 'fun': lambda w: w.sum() - 1}
```

```python
# Portafolio EMV
emv = minimize(
    fun=menos_rs,
    x0=w0,
    args=(rf, eind, sigma),
    bounds=bnds,
    constraints=cons
)
```

```python
wemv_4 = emv.x
wemv_4
```

```python
# Pesos, rendimiento, riesgo y razón de Sharpe del portafolio EMV
e_emv_4 = wemv_4.dot(eind)
s_emv_4 = np.sqrt(wemv_4.dot(sigma).dot(wemv_4))
sharpe_emv_4 = (e_emv_4 - rf) / s_emv_4
e_emv_4, s_emv_4, sharpe_emv_4
```

```python
# Comparación
print('Asi se con optimizado el sharpe')
print(e_emv_4, s_emv_4, sharpe_emv_4)

print('Asi se va con minima varianza')
print(e_mv, s_mv, sharpe_mv)
```

Además, se considerará ampliamente también si tienen en cuenta los precios reales de los activos en el momento de la compra, y como modifica esto las ponderaciones que encontraron para sus portafolios, considerando que no pueden comprar fracciones de activos y/o ETFs.

```python
# Importar librerías
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random
%matplotlib inline
```

```python
# Función para descargar precios de cierre ajustados:
def get_adj_closes(tickers, start_date=None, end_date=None, freq='d'):
    closes = web.YahooDailyReader(symbols=tickers, start=start_date, end=end_date, interval=freq).read()['Adj Close']
    closes.sort_index(inplace=True)
    return closes
```

```python
# Descargamos datos de precios de acciones
name = ['VTI','VGT', 'VCR', 'VFH', 'VHT', 
      'AAPL', 'AMZN', 'GOOGL', 'COST', 'MSFT', 
      'TSLA', 'META', 'TWTR', 'DIS', 'NFLX']
# Fechas de interés (inicio y fin)
start = '2015-01-01'
end = '2022-01-01'
# Función DataReader
closes = get_adj_closes(tickers=name,
                    start_date=start,
                    end_date=end)
```

```python
closes.head()
```

```python
# Graficamos
closes.plot(figsize = (10,7));
```

```python
# Calcular rendimientos diarios y graficarlos
ret = closes.pct_change().dropna()
ret.plot()
```

```python
# Calculamos media y desviación estándar
mu = ret.mean()
sigma = ret.std()
mu, sigma
```

```python
# Función que simula varios escenarios de rendimientos diarios
def ret_sim(mu, sigma, n_days, n_esc, start):
    dates = pd.date_range(start=start, periods=n_days)
    return pd.DataFrame(data=sigma * np.random.randn(n_days, n_esc) + mu, index=dates)
```

```python
n_esc = 30
n_days = 255

for i in name:
    # Simulamos 30 escenarios para todo el 2022
    sim_ret = ret_sim(mu[i], sigma[i], n_days, n_esc, start='2022-01-01')
    # Obtenemos los precios
    ultimo_precio = closes.iloc[-1][i]
    sim_closes = ultimo_precio * (1 + sim_ret).cumprod()
    # Concatenar y graficar
    sim_closes_cat = pd.concat([closes.iloc[-255:][i], sim_closes])
    sim_closes_cat.plot(figsize = (10,7),legend=False, title= i, xlabel='Tiempo', ylabel='Precio',  fontsize=10)
    plt.xlabel('Tiempo', fontsize=20)
    plt.ylabel('Precio', fontsize=20)
    plt.title(i, fontsize=20)
```
